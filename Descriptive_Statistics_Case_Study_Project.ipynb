```python
#Scatter plot between Employee Age and Employee Salary
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df,x='age',y='salary',color='blue',edgecolor='linen',alpha=0.5)
plt.title('Sactter Plot of employee age v/s Employee Salary')
plt.xlabel('Employee age')
plt.ylabel('Employee Salary')
plt.show()
```

Observations


*   There is a relationship between age and salary of employee
*  As Salary increases age also increases
There are also presence of outliers in the dataset



```python
#Scatter plot between Healthy Eating and Active Lifestyle
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df,x='healthy_eating',y='active_lifestyle',color='red',edgecolor='linen',alpha=0.5)
plt.title('Sactter Plot of Healthy Eating v/s Active Lifestyle')
plt.xlabel('Healthy Eating')
plt.ylabel('Active Lifestyle')
plt.show()
```

Observations



* There is a relationship between Healthy Eating and Active Lifestyle
* As Healthy Eating increases Active Lifestyle also increases




```python
# scatter plot between healthy eating and salary
#Scatter plot between Healthy Eating and Active Lifestyle
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df,x='healthy_eating',y='salary',color='k',edgecolor='linen',alpha=0.5)
plt.title('Sactter Plot of Healthy Eating v/s salary')
plt.xlabel('Healthy Eating')
plt.ylabel('salary')
plt.show()
```

Observations
* There is a relationship between HealthyEating and Salary
* As Healthy Eating increases Salary also increases



```python
#countplot with groups feature
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='groups',edgecolor='linen',alpha=0.7)
plt.title('Count Plot of Groups')
plt.xlabel('groups')
plt.ylabel('count')
plt.show()

```

Observation


*   Most Employeees Either belong to Group A or O
*  Group A has maximum frequency



```python
#histogram of salary feature
plt.figure(figsize=(11,6))
sns.histplot(data=df,x='salary',color='purple',edgecolor='linen',alpha=0.7,bins=10)
plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('count')
plt.show()
```

Observations

*  Salaries of employees are not Uniform
*  But there are employees with salaries at either extremities.And count of Highest earning employees are very less
